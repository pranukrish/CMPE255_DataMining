#Implementing Guassian Mixture model
It is probabilistic model that assumes all the data points are generated from a mixture of a finite number of Gaussian distributions with unknown parameters.


```python
#Import Libraries needed
from sklearn.mixture import GaussianMixture
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create a sample dataset
np.random.seed(0)
X = np.vstack([np.random.normal(0, 1, (1000, 4)), np.random.normal(5, 1, (1000, 4))])
```

```python

# fit a GMM model with 2 components
gmm = GaussianMixture(n_components=4)
gmm.fit(X)
```

```python
# predict the labels for the data points
labels = gmm.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis');
```

```python
probs = gmm.predict_proba(X)
print(probs[:5].round(3))
```

```python
size = 50 * probs.max(1) ** 2  # square emphasizes differences
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=size);
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_moons

# generate the moons dataset
X, y = make_moons(n_samples=500, noise=0.05)

# fit a Gaussian mixture model with 2 components to the data
gmm = GaussianMixture(n_components=2)
gmm.fit(X)

# predict the cluster labels for the data
y_pred = gmm.predict(X)

# plot the data and predicted cluster labels
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```
